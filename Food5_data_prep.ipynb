```python
import os
import cv2
from keras.preprocessing.image import load_img, img_to_array
import numpy as np
import pandas as pd
```

```python
a = "/content/drive/MyDrive/food10"
```

```python
categories = ['burger', 'jalebi', 'momos', 'paani_puri', 'pizza']
```

```python
training = []
image_size = 100
def create_train_data():
  c= 0
  for cat in categories:
    class_num = categories.index(cat)
    b = os.path.join(a,cat)
    for img in os.listdir(b):
      try:
        image = load_img(os.path.join(b,img))
        img_array = img_to_array(image)
        resized = cv2.resize(img_array,(image_size,image_size))
        training.append([resized,class_num])
      except Exception as e:
        pass
    c= c+1
    print(c)       
```

```python
create_train_data()
```

```python
training
```

```python
import random
random.shuffle(training)
```
